```
 def step_function(x):
    return np.array(x>0, dtype=np.int) # x가 0보다 크면 참(1), 0보다 작으면 거짓(0)
```

## Sigmoid
A sigmoid function is a mathematical function having a characteristic  "S"-shaped curve or sigmoid curve. 

**굵은 텍스트**$f(x) = \cfrac{1}{1+e^{-x}}$

fraction = 분수라는 뜻

```
def sigmoid(x):
    return 1 / (1 +np.exp(-x))
```

```
x = np.arange(-5.0, 5.0, 0.1)
y2 = sigmoid(x)
```

c1의 숫자가 낮으면 직선의 형태를 띄고, 숫자가 높을수록 경사가 급해져서 계단모양처럼 된다.

```
plt.plot(x, y2)
plt.grid()
plt.show()
```

$f(x) = \cfrac{1}{1+e^{-c_{1}x}}$

```
def sigmoid2(x, c1=1):
    return 1 / (1 +np.exp(-c1*x))
```

```
y2 = sigmoid2(x, 5)
```

```
plt.plot(x, y2)
plt.grid()
plt.show()
```

```
plt.plot(x, y2)
plt.grid()
plt.show()
```

```
for i in range(10):
  c1 = 0.2 * (i+1)
  y2 = sigmoid2(x, c1)
  plt.plot(x, y2)
  plt.grid()
```

* 보다 일반적인 form

$f(x) = \cfrac{1}{1+e^{-c_{1}(x-c_{2})}}$

x에 수를 빼고 더하면 그래프가 좌우로 움직임

```
def sigmoid3(x, c1=1, c2 = 0):
    return 1 / (1 +np.exp(-c1*(x - c2)))
```

```
y3 = sigmoid3(x)
```

```
plt.plot(x, y3)
plt.grid()
plt.show()
```

```
for i in range(10):
  c2 = 0.5 * i
  c1 = 2
  
  y3 = sigmoid3(x, c1, c2)
  plt.plot(x, y3)
plt.grid()
```

c2 입력하지 않으면 c2 = 0 로 되어 똑같은 위치에 그래프들 나옴
